```r
library(spacexr)
library(Matrix)
library(dplyr)
```

```r
data_dir <- "../data"

features.path <- paste0(data_dir, "/features.tsv.gz")
matrix.path <- paste0(data_dir, "/matrix.mtx.gz")
```

```r
coords <- read.csv(file.path(data_dir,"211004_02.csv"))
# coords
```

```r
coords_filt <- select(coords, c(1,3,4))
rownames(coords_filt) <- coords_filt[,1]; coords_filt[,1] <- NULL
# coords_filt
```

```r
count_mat <- readMM(file = matrix.path)
```

```r
feature.names = read.delim(features.path,
                           header = FALSE,
                           stringsAsFactors = FALSE)
rownames(count_mat) = feature.names$V1
```

```r
nUMI <- colSums(count_mat)
nUMI
```

```r
puck <- SpatialRNA(coords, count_mat, nUMI)
```

```r
## Examine SpatialRNA object (optional)
print(dim(puck@counts)) # observe Digital Gene Expression matrix
hist(log(puck@nUMI,2)) # histogram of log_2 nUMI
```

```r
print(head(puck@coords)) # start of coordinate data.frame
barcodes <- colnames(puck@counts) # pixels to be used (a list of barcode names). 
```

```r
plot_puck_continuous(puck, barcodes, puck@nUMI, ylimit = c(0,round(quantile(puck@nUMI,0.9))), title ='plot of nUMI') 
```
