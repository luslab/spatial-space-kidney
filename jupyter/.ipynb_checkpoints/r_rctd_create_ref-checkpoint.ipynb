```r
library(spacexr)
library(Matrix)
library(plyr)
library(dplyr)
```

```r
data_dir <- "../data"
```

```r
counts <- read.csv(file.path(data_dir,"Mouse_Renal_Tubule_RNA-seq_Replicates_Counts.csv"))
# counts
```

```r
dup_counts <- ddply(counts,"Gene_symbol",numcolwise(sum))
# dup_counts
```

```r
rownames(dup_counts) <- dup_counts[,1]; dup_counts[,1] <- NULL # Move first column to rownames
# dup_counts
```

```r
round_dup_counts <- ceiling(dup_counts)
round_dup_counts
```

```r
int_round_dup_counts <- sapply( round_dup_counts, as.integer ) %>% as.data.frame()
rownames(int_round_dup_counts) <- rownames(round_dup_counts)
# int_round_dup_counts
```

```r
meta_data <- read.csv(file.path(data_dir,"meta_data.csv"))
# meta_data
```

```r
cell_types <- meta_data$cluster; names(cell_types) <- meta_data$barcode # create cell_types named list
cell_types
```

```r
cell_types <- as.factor(cell_types) # convert to factor data type
cell_types
```

```r
reference <- Reference(int_round_dup_counts, cell_types)
```

```r
print(dim(reference@counts))
```

```r
table(reference@cell_types)
```

```r
saveRDS(reference, file.path(data_dir,'scref.rds'))
```
